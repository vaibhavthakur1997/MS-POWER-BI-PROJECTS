```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5 ]
y = [10,20,25,35,45]
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize =(4,3))

plt.plot(x,y, color = 'blue', marker='o', linestyle='--', linewidth=2, markersize=12)

plt.title("Sales Growth")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [5,15,25,35,45]

plt.plot(x, y1, label='Sales 2024')
plt.plot(x, y2, label='Sales 2025')

plt.title("YOY Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
# data
x = ['A','B','C','D','E']
y = [10, 20, 55, 5, 45]

plt.bar(x,y)
plt.title('Bar Chart Example')
plt.show()
```

```python
import random
data = [random.randint (1,20) for _ in range(500)]

plt.hist(data, bins = 20)
plt.title("Histogram Examples")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
Sales = [10, 20, 55, 5, 45]

plt.pie(Sales, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

```python
# used to find relationship btwn variables

# data
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.scatter(y1, y2)
plt.title("Scatter Plot Example")
plt.show()
```

```python
# used to show multiple charts in one figure

# data-1 - bar chart
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [10, 20, 55, 35, 45]

# data-2 - scatter plot
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.figure(figsize=(10,4))

# first plot- bar chart
plt.subplot(1,2,1) # row, column, position
plt.bar(categories, sales)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")


# second plot- scatter chart
plt.subplot(1,2,2) # row, column, position
plt.scatter(y1, y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")

plt.show()
```
